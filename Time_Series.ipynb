```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("http://cli.met.ie/cli/climate_data/webdata/hly4935.csv", skiprows=23, low_memory=False, nrows=1000)
```

```python
df
```

```python
pd.to_datetime(df['date'])
```

```python
#Adds column from date and converts it to a time column
df['datetime'] = pd.to_datetime(df['date'])
```

```python
df
```

```python
sns.lineplot(x="datetime", y="temp", data=df)
```

```python
#Remove or start from the row indexed by 1
sns.lineplot(x="datetime", y="temp", data=df[1:])
```

```python
import pandas as pd
```

```python
#Ref: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.date_range.html?highlight=date_range
rng = pd.date_range(start='1/1/2011', periods=72, freq='H')
rng
```

```python
import numpy as np
```

```python
ts = pd.DataFrame(np.random.randn(len(rng)), index=rng, columns=["adminssions"])
```

```python
ts
```

```python
np.random.poisson(5, 10)
```

```python
# We can use the poisson distrubution instead of the random.randn function
ts = pd.DataFrame(np.random.poisson(10, len(rng)), index=rng, columns=["adminssions"])
```

```python
ts
```

## Accessing and grouping

```python
ts.mean()
```

```python
ts.loc['2011-01-01 03:00:00']
```

```python
ts.loc['2011-01-01 03:00:00' : '2011-01-01 23:00:00']
```

```python
ts.loc['2011-01-01 03:00:00' : '2011-01-01 23:00:00'].mean()
```

```python
# Resample the dataset by day
ts.resample("D")
```

```python
# Resample the dataset by day
ts.resample("D").mean()
```
